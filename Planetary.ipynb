```python
df=pd.read_csv('Planetary Data 4y.csv')
df.head()
```

```python
moon=ephem.Moon()
sun=ephem.Sun()
mars=ephem.Mars()
venus=ephem.Venus()
mercury=ephem.Mercury()
jupiter=ephem.Jupiter()
saturn=ephem.Saturn()
uranus=ephem.Uranus()
neptune=ephem.Neptune()

deimos=ephem.Deimos()
phobos=ephem.Phobos()
ganymede=ephem.Ganymede()
titan=ephem.Titan()
pluto=ephem.Pluto()
```

```python
obs=ephem.Observer()
obs.lon = np.deg2rad(-40.9006)
obs.lat = np.deg2rad(174.8860)
```

```python
moon_alt=[]
moon_az=[]
moon_illum=[]
moon_dist=[]

sun_alt=[]
sun_az=[]
sun_dist=[]

mercury_alt=[]
mercury_az=[]
mercury_dist=[]
mercury_illum=[]

venus_alt=[]
venus_az=[]
venus_dist=[]
venus_illum=[]

mars_alt=[]
mars_az=[]
mars_dist=[]
mars_illum=[]

jupiter_alt=[]
jupiter_az=[]
jupiter_dist=[]
jupiter_illum=[]

saturn_alt=[]
saturn_az=[]
saturn_dist=[]
saturn_illum=[]

uranus_alt=[]
uranus_az=[]
uranus_dist=[]
uranus_illum=[]

neptune_alt=[]
neptune_az=[]
neptune_dist=[]
neptune_illum=[]

deimos_alt=[]
deimos_az=[]
deimos_dist=[]
deimos_illum=[]

phobos_alt=[]
phobos_az=[]
phobos_dist=[]
phobos_illum=[]

ganymede_alt=[]
ganymede_az=[]
ganymede_dist=[]
ganymede_illum=[]

titan_alt=[]
titan_az=[]
titan_dist=[]
titan_illum=[]

pluto_alt=[]
pluto_az=[]
pluto_dist=[]
pluto_illum=[]

for i in range(0,len(df)):
    obs.date=df.date[i]
    moon.compute(obs)
    moon_alt.append(moon.alt)
    moon_az.append(moon.az)
    moon_illum.append(moon.phase)
    moon_dist.append(moon.earth_distance)
    
    sun.compute(obs)
    sun_alt.append(sun.alt)
    sun_az.append(sun.az)
    sun_dist.append(sun.earth_distance)
    
    mercury.compute(obs)
    mercury_alt.append(mercury.alt)
    mercury_az.append(mercury.az)
    mercury_dist.append(mercury.earth_distance)
    mercury_illum.append(mercury.phase)
    
    venus.compute(obs)
    venus_alt.append(venus.alt)
    venus_az.append(venus.az)
    venus_dist.append(venus.earth_distance)
    venus_illum.append(venus.phase)
    
    mars.compute(obs)
    mars_alt.append(mars.alt)
    mars_az.append(mars.az)
    mars_dist.append(mars.earth_distance)
    mars_illum.append(mars.phase)
    
    jupiter.compute(obs)
    jupiter_alt.append(jupiter.alt)
    jupiter_az.append(jupiter.az)
    jupiter_dist.append(jupiter.earth_distance)
    jupiter_illum.append(jupiter.phase)
    
    saturn.compute(obs)
    saturn_alt.append(saturn.alt)
    saturn_az.append(saturn.az)
    saturn_dist.append(saturn.earth_distance)
    saturn_illum.append(saturn.phase)
    
    uranus.compute(obs)
    uranus_alt.append(uranus.alt)
    uranus_az.append(uranus.az)
    uranus_dist.append(uranus.earth_distance)
    uranus_illum.append(uranus.phase)
    
    neptune.compute(obs)
    neptune_alt.append(neptune.alt)
    neptune_az.append(neptune.az)
    neptune_dist.append(neptune.earth_distance)
    neptune_illum.append(neptune.phase)
    
    deimos.compute(obs)
    deimos_alt.append(neptune.alt)
    deimos_az.append(neptune.az)
    deimos_dist.append(neptune.earth_distance)
    deimos_illum.append(neptune.phase)
    
    phobos.compute(obs)
    phobos_alt.append(neptune.alt)
    phobos_az.append(neptune.az)
    phobos_dist.append(neptune.earth_distance)
    phobos_illum.append(neptune.phase)
    
    ganymede.compute(obs)
    ganymede_alt.append(neptune.alt)
    ganymede_az.append(neptune.az)
    ganymede_dist.append(neptune.earth_distance)
    ganymede_illum.append(neptune.phase)
    
    titan.compute(obs)
    titan_alt.append(neptune.alt)
    titan_az.append(neptune.az)
    titan_dist.append(neptune.earth_distance)
    titan_illum.append(neptune.phase)
    
    pluto.compute(obs)
    pluto_alt.append(neptune.alt)
    pluto_az.append(neptune.az)
    pluto_dist.append(neptune.earth_distance)
    pluto_illum.append(neptune.phase)
```

```python
df["Sun_alt"]=sun_alt
df["Sun_az"]=sun_az
df["Sun_dist"]=sun_dist

df["Moon_alt"]=moon_alt
df["Moon_az"]=moon_az
df["Moon illum"]=moon_illum
df["Moon_dist"]=moon_dist

df["Mercury_alt"]=mercury_alt
df["Mercury_az"]=mercury_az
df["Mercury_dist"]=mercury_dist
df["Mercury_illum"]=mercury_illum

df["Venus_alt"]=venus_alt
df["Venus_az"]=venus_az
df["Venus_dist"]=venus_dist
df["Venus_illum"]=venus_illum

df["Mars_alt"]=mars_alt
df["Mars_az"]=mars_az
df["Mars_dist"]=mars_dist
df["Mars_illum"]=mars_illum

df["Jupiter_alt"]=jupiter_alt
df["Jupiter_az"]=jupiter_az
df["Jupiter_dist"]=jupiter_dist
df["Jupiter_illum"]=jupiter_illum

df["Saturn_alt"]=saturn_alt
df["Saturn_az"]=saturn_az
df["Saturn_dist"]=saturn_dist
df["Saturn_illum"]=saturn_illum

df["Uranus_alt"]=uranus_alt
df["Uranus_az"]=uranus_az
df["Uranus_dist"]=uranus_dist
df["Uranus_illum"]=uranus_illum

df["Neptune_alt"]=neptune_alt
df["Neptune_az"]=neptune_az
df["Neptune_dist"]=neptune_dist
df["Neptune_illum"]=neptune_illum

df["Deimos_alt"]=neptune_alt
df["Deimos_az"]=neptune_az
df["Deimos_dist"]=neptune_dist
df["Deimos_illum"]=neptune_illum

df["Phobos_alt"]=phobos_alt
df["Phobos_az"]=phobos_az
df["Phobos_dist"]=phobos_dist
df["Phobos_illum"]=phobos_illum

df["Ganymede_alt"]=ganymede_alt
df["Ganymede_az"]=ganymede_az
df["Ganymede_dist"]=ganymede_dist
df["Ganymede_illum"]=ganymede_illum

df["Titan_alt"]=titan_alt
df["Titan_az"]=titan_az
df["Titan_dist"]=titan_dist
df["Titan_illum"]=titan_illum

df["Pluto_alt"]=pluto_alt
df["Pluto_az"]=pluto_az
df["Pluto_dist"]=pluto_dist
df["Pluto_illum"]=pluto_illum
```
